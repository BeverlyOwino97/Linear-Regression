```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
```

```python
df
```

```python
df2 = df.rename(columns = {'per capita income (US$)': 'Income'}, inplace = False)
```

```python
df2
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('Income')
plt.scatter(df2.year,df2.Income,color='red',marker='+')
```

```python
df_mpya = df2.drop('Income', axis = 'columns')
df_mpya
```

```python
Income = df2.Income
Income
```

```python
# Create linear regression object
model = linear_model.LinearRegression()
model.fit(df_mpya,Income)
```

```python
model.predict([[2017]])
```

```python
model.coef_
```

```python
model.intercept_
```
