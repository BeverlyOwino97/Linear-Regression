```python
import matplotlib.pyplot as plt
def gradient_descent(x,y):
    w1_curr = w0_curr = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.006
    for i in range(iterations):
        y_predicted = w1_curr*x + w0_curr
        cost = (1/n)*sum([val**2 for val in (y-y_predicted)])
        w1d = -(2/n)*sum(x*(y-y_predicted))
        w0d = -(2/n)*sum(y-y_predicted)
        w1_curr = w1_curr - (learning_rate*w1d)
        w0_curr = w0_curr - (learning_rate*w0d)
        print("w1{}, wo{}, cost{}, iteration{}".format(w1_curr, w0_curr, cost, i))
        
    plt.plot(x, y_predicted, 'r')
x = np.linspace(1,10,20)
noise = np.random.randn(20)
y = 2*x +1 + noise
gradient_descent(x,y)
plt.scatter(x,y)
plt.grid()
```

```python
# mean of x and y vector
mean_x = x.mean()
mean_y = y.mean()
# calculating cross-deviation and deviation about x
sum_yx = (x*y).sum()
x_sq = (x**2).sum()
ssxy = sum_yx - (len(x)*mean_x*mean_y)
ssxx = ((x - mean_x)**2).sum()
# calculating regression coefficients
theta_1 = ssxy/ssxx
theta_0 = mean_y - (theta_1*mean_x)
# printing both the values
print('Theta 0: {:0.3f}'.format(theta_0))
print('Theta 1: {:0.3f}'.format(theta_1))
```
